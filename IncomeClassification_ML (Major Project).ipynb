```python
columns=['Age','Workclass','Fnlwgt','Education','Education_num','Marital_status','Occupation','Relationship','Race','Sex','Capital_gain','Capital_loss','Hours_per_week','Native_country','Income']
df=pd.read_csv('adult.csv',header=None,na_values=[" ?"],names=columns)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
my_df = df.dropna()
```

```python
my_df.info()
```

```python
my_df.isnull().sum()
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import LabelEncoder
my_df = my_df.apply(LabelEncoder().fit_transform)
my_df.head()
```

```python
drop_elements = ['Education', 'Native_country', 'Income']
y = my_df["Income"]
x = my_df.drop(drop_elements, axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
from sklearn.metrics import precision_score, accuracy_score
from sklearn.metrics import recall_score, f1_score
```


# Decision Tree
---



```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(criterion='gini',max_depth=10,min_samples_split=25)
tree.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

print('Tree-training score: ',tree.score(x_train,y_train))
print('Tree-testing score: ',tree.score(x_test,y_test))
```

```python
dt_predict = tree.predict(x_test)
cm_dt = confusion_matrix(y_test,dt_predict)
cr_dt = classification_report(y_test,dt_predict)
print('Confusion matrix\n',cm_dt)
print('\n')
print('Classifiation report\n',cr_dt)

print('\nPrecision: ', precision_score(y_test, dt_predict))
print('Recall: ', recall_score(y_test, dt_predict))
print('F1 Score: ', f1_score(y_test, dt_predict))
print("Accuracy Rate: ", accuracy_score(y_test,dt_predict))
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 8)
knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

print('KNN-training score: ',knn.score(x_train,y_train))
print('KNN-testing score: ',knn.score(x_test,y_test))
```

```python
knn_predict = knn.predict(x_test)
```

```python
cm_knn = confusion_matrix(y_test,knn_predict)
cr_knn = classification_report(y_test,knn_predict)
print('Confusion Matrix\n',cm_knn)
print('\n')
print('Classification Report\n',cr_knn)

print('\nPrecision: ', precision_score(y_test, knn_predict))
print('Recall: ', recall_score(y_test, knn_predict))
print('F1 Score: ', f1_score(y_test, knn_predict))
print("Accuracy Rate: ", accuracy_score(y_test,knn_predict))
```
